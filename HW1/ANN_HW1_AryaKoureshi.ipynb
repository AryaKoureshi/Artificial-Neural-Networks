برخلاف نگاه مک‌لوچ‌پیتز، نگاه هب به نورون مدلسازی شده، نگاهی ازمنظر جبر بولی نبود، بلکه بیشتر به مشاهدات زیست‌شناسان همان زمان به نورون‌های واقعی بود. هب فرض کرد که یا اتفاقی خواهد افتاد، یا نخواهد افتاد. درحالی‌که نگاه جبر بولی، 0 و 1 به معنای رخداد و عدم رخداد نیستند. بلکه هر دو مقدار قراردادی برای نمایش یک رخداد محسوب می‌شوند. در مقابل این نگاه جبر بولی، زیست‌شناسان میگفتند که اگر ورودی به آستانه‌ای برسد، رخداد خروجی حاصل خواهد شد و به اصطلاح نورون اسپایک خواهد زد. قانون یا همه یا هیچ برقرار است و برای اسپایک زدن نورون، ورودی باید به یک آستانه‌ای برسد در غیر اینصورت اسپایک نخواهیم داشت. مک‌لوچ‌ و پیتز به علت بستر پردازشی جبر بولی، به سراغ آن رفتند و نگاه جبری داشتند و به ورودی‌ها و خروجی‌ها مقادیر عددی نسبت دادند. اما هب به همان نگاه اولیه وجود یا عدم وجود برگشت. این دلیلی بود که هب از رابطه delta = x.y استفاده کرد که بیانگر وابستگی تغییرات وزن‌ها به رخداد وجود خروجی است.
زمانی که از ورودی‌ها و خروجی‌های باینری استفاده می‌کردیم، یادگیری فقط زمانی اتفاق می‌افتاد که هم خروجی و هم ورودی‌ها فعال باشند یعی مقدار 1 داشته باشند. اما اگر این فضا که دو حالت دارد را از کدینگ باینری به کدینگ بایپولار نگاشت کنیم، یادگیری جای یک حالت، دو حالت خواهد داشت به اینصورت که یک حالت زمانی است که مانند قبل هم خروجی و هم ورودی فعال باشند (1 باشند) و حالت دیگر زمانی است که هر دو هم خروجی و هم ورودی غیرفعال باشند (مقدار 1- داشته باشند). 
علاقه‌ی ما بیشتر به سمت فضای بایپولار است چون در فضای باینری حالت صفر همواره باعث نتیجه‌ای تحت عنوان عدم یادگیری خواهد شد که میزان کارکرد مدل را کاهش خواهد داد.
با تغییر فضا از باینری به بایپولار، میزان یادگیری تقویت می‌شود. اما این به معنای اثبات آنکه این مدل یادگیری کامل را دارد، نیست. به عبارتی گرچه پس از مطرح شدن مدل هب امیدهایی در زمنیه‌ی مدلسازی مغز پدید آمد، اما به زودی معلوم شد، مصادیقی برای نقض یادگیری مدل هب وجود دارد؛ حتی در موارد خطی تفکیک پذیر. پس از آن پرسشی به وجود آمد که آیا می‌توان دسته‌بندی خاصی برای مسائلی که قابل یادگیری یا غیرقابل یادگیری با مدل هب هستند ارائه داد؟ پاسخ به این سوال خیر بود و معلوم شد که هیچ تضمینی برای آنکه آیا مدل هب قادر به یادگیری یک فانکشن هست یا نه، نمی‌توان ارائه داد. یعنی نمی‌توان تضمین نمود که همگرا خواهد شد. به همین دلیل در این گزارش بارها با کلماتی مبنی بر عدم قطعیت مواجه خواهیم شد. 
نکته دیگر آن است که در صورت رخداد خطا و خراب شدن بعضی مولفه‌ها، مدل هب تا زمانی می‌تواند این خطا را در نظر نگیرد که حالت ایجاد شده از فصل مشترک افراز کننده‌ی فضا (رابطه خطی حاصله از مدل‌سازی هب)، عبور نکرده و در طرف دیگر فصل مشترک قرار نگیرد. در این صورت، کاملاً مشهود است که شبکه‌های عصبی به نوعی توانایی تعمیم‌پذیری دارند؛ به عبارتی به ازای یک پترن ورودی، الزاماً فقط آن پترن را تشخیص نخواهند داد. بلکه فضایی شامل آن پترن را تشخیص خواهند داد که این امکانات خوبی همچون حذف نویز را خواهد داد.

##### <div dir="rtl"> در مورد فضای ویژگی توضیح دهید و بگویید انتظار دارید در صورتی که پترن‌های در نظر گرفته شده متعامد باشند چه اتفاقی رخ بدهد، و چه تاثیری در یادگیری خواهد داشت؟ </div>

<div dir="rtl">
فضای ویژگی یک فضای چند بعدی است که برای نمایش ویژگی‌های مختلف یک داده استفاده می‌شود. به عبارت دیگر، فضای ویژگی یک فضایی است که هر نقطه در آن نمایانگر یک داده با ویژگی‌های مختلف است. در فضای ویژگی، داده‌ها با استفاده از ترکیب خطی از ویژگی‌ها نمایش داده می‌شوند. برای مثال، اگر ویژگی‌های یک داده شامل طول، عرض و ارتفاع باشد، داده مربوطه می‌تواند در فضای سه بعدی نمایش داده شود، که هر محور آن نمایانگر یکی از ویژگی‌هاست.
در صورتی که پترن‌های در نظر گرفته شده متعامد باشند، به این معنی است که هیچ ارتباط خطی بین آن‌ها وجود ندارد و هیچ رابطه‌ی خطی بین ویژگی‌ها وجود ندارد. در این حالت، داده‌ها به شکلی در فضای ویژگی قرار می‌گیرند که احتمالا به نظر می‌رسد که هیچ ساختاری در داده وجود ندارد و به نظر می‌رسد که داده‌ها به شکل تصادفی در فضای ویژگی قرار گرفته‌اند.
به طور کلی، وجود پترن‌های متعامد در فضای ویژگی می‌تواند تاثیر مخربی بر روی یادگیری داشته باشد. زیرا در این حالت، الگوریتم‌های یادگیری نمی‌توانند الگوهایی در داده‌ها را شناسایی کنند. به عبارت دیگر، در فضای ویژگی به شکلی قرار می‌گیرند که هیچ رابطه‌ی خطی بین ویژگی‌ها وجود ندارد و هیچ الگویی در داده‌ها وجود ندارد. در نتیجه، یادگیری با مشکلاتی روبه‌رو خواهد شد و به شدت مشکل خواهد بود. اما این نکته را نیز باید در نظر داشت که بسته به نوع الگوریتم یادگیری و ویژگی‌های مورد استفاده، تاثیر وجود پترن‌های متعامد در فضای ویژگی ممکن است کمتر یا بیشتر باشد.
